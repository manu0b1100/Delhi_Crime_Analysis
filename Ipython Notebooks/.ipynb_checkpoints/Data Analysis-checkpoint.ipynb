```python
import pandas as pd
```

```python
df=pd.read_csv("Dataset2/charged.csv")
```

```python
df
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
df=df[~(df.Long==df.Long.max()).all() and ~(df.Lat==df.Lat.max())]
```

```python
m= Basemap(projection='mill', llcrnrlat=28.423229, urcrnrlat=28.879376,
                    llcrnrlon=76.810895, urcrnrlon=77.348533, resolution='c', epsg=4269)
# m = Basemap(projection='mill', llcrnrlat=df.Lat.max(), urcrnrlat=df.Lat.min(),
#                     llcrnrlon=df.Long.min(), urcrnrlon=df.Long.max(), resolution='c', epsg=4269)
x, y = m(tuple(df.Long[(df.Long.isnull() == False)]), tuple(df.Lat[(df.Lat.isnull() == False)]))
plt.figure(figsize=(40, 20))
m.arcgisimage(service="NatGeo_World_Map", xpixels=2000,verbose=True)
m.plot(x, y, 'ro', markersize=8, alpha=1,color='red')
plt.show()
```
