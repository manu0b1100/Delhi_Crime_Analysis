```python
counts=df.datefrom.value_counts()
dates=counts[counts>10].index.tolist()
l=len(dates)
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

fig = plt.gcf()
plt.figure(figsize=(10,9))
m= Basemap(projection='mill', llcrnrlat=28.423229, urcrnrlat=28.879376,llcrnrlon=76.810895, urcrnrlon=77.348533, resolution='c', epsg=4269)
x, y = m(df.Long[(df.Long.isnull() == False)], df.Lat[(df.Lat.isnull() == False)])
m.arcgisimage(service="NatGeo_World_Map", xpixels=2000,verbose=True)
a, b = m(df.Long, df.Long)
graph = m.plot(a, b, 'ro', markersize=6, alpha = 1.0)[0]
```

```python
from matplotlib.animation import FuncAnimation

def animate(i,dates):
    lons = x[df.datefrom==dates[i]]
    lats = y[df.datefrom==dates[i]]
    graph.set_data(lons, lats)
    return graph

anim = FuncAnimation(plt.gcf(), animate, repeat=True, frames=l, interval=500,fargs=(dates,))
```
